### Clean Student `CommentTags`

```python
comTags = df['CommentTags']
comTags
```

```python
from string import punctuation

def remove_punctuation(document):
    keep_comma = punctuation.replace(',', '')
    no_punct = ''.join([character for character in document if character not in keep_comma])
    
    return no_punct
```

#### Remove punctuation, but keep comma for further separation

```python
comTags_no_punct = comTags.apply(remove_punctuation)
comTags_no_punct
```

#### Lowercase

```python
comTags_lower = comTags_no_punct.apply(str.lower)
comTags_lower
```

#### Turn combined string into separate feature tags

```python
def process_tag(tags):
    tags_list = tags.split(',')
    cleaned_tags = []
    
    for item in tags_list:
        item = item.strip()
        item = item.replace(' ', '_')
        cleaned_tags.append(item)
    
    return cleaned_tags 
```

```python
comTags_cleaned = comTags_lower.apply(process_tag)
comTags_cleaned
```

```python
tag_set = set([])
for index, text_list in comTags_cleaned.items():
    tag_set.update(text_list)
```

```python
tag_set
```

```python
len(tag_set)
```

```python
tag_df = pd.DataFrame(columns = tag_set)
tag_df = tag_df.sort_index(axis = 1)
tag_df
```

```python
tag_data = np.zeros((len(comTags_cleaned), len(tag_set)))

for index, tag_list in comTags_cleaned.items():
    for tag in tag_list:
        if tag in list(tag_df.columns):
            col_index = list(tag_df.columns).index(tag)
            tag_data[index, col_index] = 1
```

```python
tag_encode = pd.DataFrame(tag_data, columns = tag_df.columns, dtype='int32')
tag_encode
```

#### Combine similar meaning attributes

```python
tag_encode['caring'] = tag_encode['caring'] + tag_encode['cares_about_students']
tag_encode['extra_credit'] = tag_encode['extra_credit'] + tag_encode['extra_credit_offered']
tag_encode['respected'] = tag_encode['respected'] + tag_encode['respected_by_students']
```

```python
tag_encode = tag_encode.drop(['cares_about_students', 'extra_credit_offered', 'respected_by_students'], axis = 1)
```

### Background

- Treat `Missing` as one explicit category

```python
bg = df_new['Background']
bg
```

```python
from itertools import repeat

def creteria_split(bg):
    criteria = ['Credit', 'Attendance', 'Again', 'Grade', 'Textbook']
    one_hot = list(repeat('Missing', 5))
    
    # first choose non-element
    text_list = re.findall('[a-zA-Z]+', bg)
    if text_list == []:
        return one_hot

    for item in text_list:
        if item in criteria:
            index = criteria.index(item)
            content_index = text_list.index(item) + 1
            one_hot[index] = text_list[content_index]

    return one_hot
```

```python
bg = bg.apply(creteria_split)
bg
```

```python
Credit_list = []
Atten_list = []
Again_list = []
Grade_list = []
Textbook_list = []

for index, creteria in bg.items():
    Credit_list.append(creteria[0])
    Atten_list.append(creteria[1])
    Again_list.append(creteria[2])
    Grade_list.append(creteria[3])
    Textbook_list.append(creteria[4])

bg_filtered = pd.DataFrame({'Credit': Credit_list,
                           "Attendance": Atten_list,
                           "Student_TakeAgain": Again_list,
                           "Grade": Grade_list,
                           "Textbook": Textbook_list})
```

```python
bg_filtered
```

#### Deal with some 'tricky' values in entries

```python
set(bg_filtered['Textbook'].values)
```

#### Treat all 'N' as 'No'

```python
N_index = []
for index, text_list in bg_filtered['Textbook'].items():
    if text_list == 'N':
        N_index.append(index)
```

```python
bg_filtered['Textbook'][N_index] = 'No'
```

```python
set(bg_filtered['Textbook'].values)
```

#### Grade

```python
set(bg_filtered['Grade'].values)
```

```python
for index, text in bg_filtered['Grade'].items():
    if text == 'Not':
        bg_filtered['Grade'][index] = 'NotSure'
    if text == 'Audit':
        bg_filtered['Grade'][index] = 'AuditNoGrade'
```

```python
set(bg_filtered['Grade'].values)
```

#### Let's look at how much entries having less than or equal to 1 `Missing` value

```python
index_list = []
for index, value_list in bg_filtered.iterrows():
    i = 0
    for value in value_list:
        if value == 'Missing':
            i += 1
    if i <= 1:
        index_list.append(index)
```
